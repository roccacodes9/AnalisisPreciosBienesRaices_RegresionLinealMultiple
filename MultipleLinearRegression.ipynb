# Multiple Linear Regression

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns

# Configuraci√≥n de seaborn para mejorar el estilo de las visualizaciones
sns.set()
```

## Load the data

```python
# Cargar los datos desde un archivo CSV
data = pd.read_csv('real_estate_price_size_year.csv')
```

```python
# Visualizar las primeras filas del DataFrame para entender su estructura
data.head()
```

```python
# Obtener estad√≠sticas descriptivas del DataFrame para comprender la distribuci√≥n de los datos
data.describe()
```

## Create the regression

### Declare the dependent and the independent variables

```python
# Definir las variables dependientes e independientes
y = data['price']
x1 = data[['size', 'year']]
```

### Regression

```python
# Agregar una columna constante para el t√©rmino de intercepci√≥n en el modelo
x = sm.add_constant(x1)

# Ajustar un modelo de regresi√≥n lineal ordinaria (OLS)
results = sm.OLS(y, x).fit()

# Mostrar un resumen detallado de los resultados del modelo
results.summary()
```
